"values" -- defintion of what are searching for

dictionaries use `{}` instead of lists using `[]`

```python
a_words = []

for item in words:
    if(item.startswith('a')):
#         print(item)
        a_words.append(item)

print(a_words)
```

```python
words = open("frost.txt").read().split()

# dictionaries are {} and we are making an empty one
initials = {}

for item in words:
    first_letter = item[0]
    first_letter = first_letter.upper()
    if first_letter not in initials:
        
        initials[first_letter] = []
    initials[first_letter].append(item)

print(initials.keys())
```

```python
print(initials.values())
```

```python
print(initials["T"])
```

```python
import random

for letter in "robertfrost":
    letter = letter.upper()
    word = random.choice(initials[letter])
    print(word[0].upper() + word[1:])
```

```python
# dictionaries are {} and we are making an empty one
pos = {}

for item in words:
    first_letter = item.pos_
    if first_letter not in pos:
        pos[first_letter] = []
    pos[first_letter].append(item)

print(pos.keys())
```

```python
import random

print(random.choice(pos['NOUN']))
```

```python
# adding the .text attribute after getting the random choice allows you to convert to a string instead of a spacy object

my_sentence = random.choice(pos['NOUN']).text + " " + random.choice(pos['VERB']).text + " "+random.choice(pos['PROPN']).text

print(my_sentence)
```
